#### Null값을 처리할 수 있는 전처리 방법을 생각해보기

1.데이터 Null값을 모두 0으로 대체하기(즉, Null값은 No Signal을 나타낸다 판단한다)

2.Correlation method를 통해 결과에 대한 Feature의 상관관계를 파악하고 상위 40개 Feature들만 사용하기

3.칼럼별 결측값 500개 이하는 평균값으로 대체, 500개 이상인 열들은 삭제하기

### Proprocessing Method 1

```python
# Replace the Null values with Zeros 
data1 = data.replace(np.NaN, 0)
```

```python
# Check up if there is any Null values left
data1.isnull().any().any()
```

```python
# Get rid of 'Time' column because it's meaningless
new_data1=data1.drop(['Time'],axis=1)
```

### Proprocessing Method 2

```python
# Find features that are ranked in the top 40 rankings by using 'Correlation Method'
corrMatrix = data.corr()
corrMatrix
```

```python
sns.heatmap(corrMatrix,0.1)
plt.show()
```

### Proprocessing Method 3

```python
# Count the number of Null values
nas=data.isnull().sum()
# Change 'nas' from Matrix form to array
nas.values
```

```python
# Exclude the column that has more than 500 Null values
data3 = data[data.columns[nas<500]]
```

```python
# Fill the Null values with the mean
data3 = data3.fillna(data3.mean())
```

```python
# Get rid of 'Time' column because it's meaningless
new_data3=data3.drop(['Time'],axis=1)
```
