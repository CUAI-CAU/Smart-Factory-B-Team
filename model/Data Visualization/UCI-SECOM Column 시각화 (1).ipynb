### UCI-Secom Column 시각화

```python
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings(action='ignore')
```

```python
# Read the data
data = pd.read_csv('uci-secom.csv')
```

편차값이 큰 것을 보이기 위한 Column: 1, 7

0으로 다 같은 고유치가 1임을 보여주기 위한 Column: 13

```python
# unique_vals로 Pass/Fail 열의 고유치 뽑아내기(-1과 1)
unique_vals = data['Pass/Fail'].unique()
unique_vals
```

```python
# Pass/Fail행의 고유치가 각각 1과 -1인 행의 데이터 뽑아내기
targets = [data.loc[data['Pass/Fail'] == val] for val in unique_vals]
targets
```

```python
fig = plt.figure(figsize=(10,10))
for target in targets:
    sns.distplot(target['1'], hist=True, rug=True)
plt.title('1st Sensor Measurements', fontsize = 20)
plt.xlabel('7', fontsize=15)
plt.ylabel('Density', fontsize=15)
```

```python
fig = plt.figure(figsize=(10,10))
for target in targets:
    sns.distplot(target['7'], hist=True, rug=True)
plt.title('7th Sensor Measurements', fontsize = 20)
plt.xlabel('7', fontsize=15)
plt.ylabel('Density', fontsize=15)
```

```python
fig = plt.figure(figsize=(10,10))
for target in targets:
    sns.distplot(target['13'], hist=True, rug=True)
plt.title('13th Sensor Measurements', fontsize = 20)
plt.xlabel('13', fontsize=15)
plt.ylabel('Density', fontsize=15)
```
